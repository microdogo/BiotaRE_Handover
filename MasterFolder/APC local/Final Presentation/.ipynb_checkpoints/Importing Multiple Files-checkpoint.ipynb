### Import Multiple CSV file

```python
import os
import glob
import pandas as pd
import numpy as np
file_path = "C:/Users/Fernando Gutierrez/Desktop/test"
read_files = glob.glob(os.path.join(file_path, "*.csv")) 
                       
np_array_values = []
for files in read_files:
    data_load = pd.read_csv(files, header = 0)
    np_array_values.append(data_load)                   
    print(files)   
merge_values = np.vstack(np_array_values)
data_loaded = pd.DataFrame(merge_values)
data_loaded.columns=['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten']
data_loaded
```

### Reservoir Modeling - APC Data

#### Scope:
- To build and compare reservoir models with and without DNA Surveillance data
- In this notebook we'll focus on the wells for which we have geology, completions, production and QC-passed DNA data (~90 wells)
- Random Forest method is used to predict 2-year cumulative oil production as a function of completions, geologic and/or DNA formation contributions data

#### Main Findings:
- Even when all available geologic and completions parameters are used (Model 1), the model doesn't score a high $R^{2}$ in 5-fold cross validation. This suggests that important features are missing from the data (e.g. natural fracture system, bottomhole flowing pressure), which can be addressed by using prior knowledge of oil production (e.g. 3-months production) to calibrate the model for e.g. 2-year production. 
- Model 3 in which DNA formation contributions are used as proxies for effective geologic properties yields comparable accuracy to Model 1 in which all geologic properties are used (**0.43 vs. 0.44** 5-fold cross-validation $R^{2}$)

#### Organizing Data:

```python
# select a subset of columns that we will be working with
X0 = df[regression_params]
```

```python
# let's convert categorical parameters to numeric (binary) values
mapping1 = {'SSLEEVE': 0, 'PNP': 1}
mapping2 = {'SLICKWATER FRAC': 0, 'CROSSLINK': 1}
X0.replace({'Liner_Type': mapping1, 'PRIMARY_FLUID':mapping2}, inplace=True)
```

#### Model 1: Using Geologic and Completions Parameters for Individual Wells

```python
# select independent variables (geologic and completion parameters)
X = X0.iloc[:,1:23]
```

```python
# convert feature data frame to a numpy array, but first save feature names for later
feature_list = list(X.columns)
X = np.array(X)
y = np.ravel(np.array(y_o))
```

```python
#lm = LinearRegression()
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=42, n_estimators=100)
```

```python
# split the data into training and test sets
X_train, X_test, y_o_train, y_o_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=True)
```

```python
# fit a model to the training data set
#model1 = lm.fit(X_train, y_o_train)
model1 = rf.fit(X_train, y_o_train)
```

```python
# R^2 for training and test set
model1.score(X_train, y_o_train), model1.score(X_test, y_o_test)
```

```python
# Make predictions using the training & testing set
y_o_pred_train = model1.predict(X_train)
y_o_pred_test = model1.predict(X_test)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))

ax.scatter(y_o_train, y_o_pred_train,  color='gray', label="Training Data")
ax.scatter(y_o_test, y_o_pred_test,  color='green', label="Testing Data")
#diag = np.linspace(0, 6e5, 10)
diag = np.linspace(11, 14, 10)
ax.plot(diag, diag, ls= "--", c="0.3")
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))    
ax.set_xlabel("Actual LogCumOil_24m")
ax.set_ylabel("Predicted LogCumOil_24m")
ax.set_title("Model 1 - without DNA")
,

fig.tight_layout()
```

Now let's perform 5-fold cross validation

#### Model 2: Genomics-Assisted, when Completions and `Average` Formation Properties are Available

Assuming we only know average formation (geologic) properties, we can use DNA formation contributions to calculate effective geologic properties in the drainage volume of induvidual wells for use in modeling.

```python
# formation contribution to each well
A = X0.iloc[:,23:28].as_matrix()
```

```python
# average geologic properties of each formation
B = X0.iloc[:,0:10].groupby('Formation').mean().as_matrix()
```

```python
# each well's effective geologic properties = weighted average based on formations contribution to that well
C = np.matmul(A,B)
```

```python
# convert C to a data frame
C = pd.DataFrame(C)
# assign index and column names
C.index = X0.index
C.rename(columns=dict(zip(C.columns[0:], X0.columns[1:10])),inplace=True)
```

```python
# place the computed properties in the regressors data frame
X2 = X0.iloc[:,1:23]
X2.iloc[:,0:9] = C
```

```python
# convert feature data frame to a numpy array, but first save feature names for later
feature_list = list(X2.columns)
X2 = np.array(X2)
y = np.ravel(np.array(y_o))
```

```python
# split the data into training and test sets
X_train, X_test, y_o_train, y_o_test = train_test_split(X2, y, test_size=0.3, random_state=42)
```

```python
# fit a model to training data set
#model2 = lm.fit(X_train, y_o_train)
model2 = rf.fit(X_train, y_o_train)
```

```python
# R^2 for training and test set
model2.score(X_train, y_o_train), model2.score(X_test, y_o_test)
```

```python
# Make predictions using the training & testing set
y_o_pred_train = model2.predict(X_train)
y_o_pred_test = model2.predict(X_test)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))

ax.scatter(y_o_train, y_o_pred_train,  color='gray', label="Training Data")
ax.scatter(y_o_test, y_o_pred_test,  color='green', label="Testing Data")
#diag = np.linspace(0, 6e5, 10)
diag = np.linspace(11, 14, 10)
ax.plot(diag, diag, ls= "--", c="0.3")
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))    
ax.set_xlabel("Actual LogCumOil_24m")
ax.set_ylabel("Predicted LogCumOil_24m")
ax.set_title("Model 2 - with DNA")
,

fig.tight_layout()
```

Now let's perform 5-fold cross validation

```python
cv_results = cross_validate(model2, X2, y, cv=5, return_train_score=False)
```
